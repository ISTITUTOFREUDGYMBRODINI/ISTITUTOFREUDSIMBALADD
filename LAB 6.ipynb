## BLINKING LED CON PWM / BLINKING LED WITH PWM

```python
import time
from pyfirmata import Arduino, PWM

# Inizializza la connessione con l'Arduino
board = Arduino('COM7')  # Sostituisci 'COM3' con la tua porta seriale corretta

# Definisci il pin a cui è collegato il LED
pin_led = board.get_pin('d:11:p')  # Utilizziamo il pin 9 come pin PWM

# Effetto di fading in
def fading_in():
    for brightness in range(0, 101):
        pin_led.write(brightness / 100.0)
        time.sleep(0.02)  # Ritardo per controllare la velocità del fading

# Effetto di fading out
def fading_out():
    for brightness in range(100, -1, -1):
        pin_led.write(brightness / 100.0)
        time.sleep(0.02)  # Ritardo per controllare la velocità del fading
        
def main():    
    for n in range(100):
        fading_in()
        fading_out()
        pin_led.write(0)
    # Chiudi la connessione con l'Arduino
    board.exit()
    

if __name__ == "__main__":
    main()

# Assicurati di spegnere il LED alla fine
```

## PUSHBUTTON CON LED / PUSHBUTTON WITH LED

```python
import time
from pyfirmata import Arduino, util

# Definisci il numero del pin per il pulsante e il LED
button_pin_num = 2
led_pin_num = 11

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Definisci il pin a cui è collegato il LED e il pulsante
led_pin = board.get_pin(f'd:{led_pin_num}:o')  # Pin digitale 11 come output
button_pin = board.get_pin(f'd:{button_pin_num}:i')  # Pin digitale 2 come input

# Avvia un iteratore per leggere i pin digitali
it = util.Iterator(board)
it.start()

# Abilita la lettura del pin del pulsante
button_pin.enable_reporting()

while True:
    # Leggi lo stato del pulsante
    button_state = button_pin.read()

    # Controlla se il pulsante è premuto
    if button_state:
        # Accendi il LED
        led_pin.write(1)
    else:
        # Spegni il LED
        led_pin.write(0)

    # Aggiungi un piccolo ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.1)

```

## PUSHBUTTON CON LED + MONITOR SERIALE / PUSHBUTTON WITH LED + SERIAL MONITOR

```python
import time
from pyfirmata import Arduino, util

# Definisci il numero del pin per il pulsante e il LED
pinButton = 2
pinLed = 11

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Configura i pin del pulsante e del LED
button_pin = board.get_pin(f'd:{pinButton}:i')  # Pin digitale 2 come input
led_pin = board.get_pin(f'd:{pinLed}:o')  # Pin digitale 11 come output

# Variabili per lo stato del pulsante, lo stato precedente e lo stato del LED
valButton = False
valButtonOld = False
ledState = False

# Avvia un iteratore per leggere i pin digitali
it = util.Iterator(board)
it.start()

while True:
    # Leggi lo stato del pulsante
    valButton = button_pin.read()

    # Aggiorna lo stato del LED se il pulsante è stato premuto
    if valButton and not valButtonOld:
        ledState = not ledState

    # Stampa lo stato del LED
    if ledState:
        print("Il LED è acceso")
    else:
        print("Il LED è spento")

    # Scrivi lo stato del LED sul pin
    led_pin.write(ledState)

    # Aggiorna lo stato precedente del pulsante
    valButtonOld = valButton

    # Aggiungi un ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.1)
```

## PUSHBUTTON + MONITOR SERIALE / PUSHBUTTON + SERIAL MONITOR

```python
import time
from pyfirmata import Arduino, util

# Definisci il numero del pin per il pulsante
buttonPin = 2

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Configura il pin del pulsante
button_pin = board.get_pin(f'd:{buttonPin}:i')  # Pin digitale 2 come input

# Variabili per lo stato del pulsante, lo stato precedente, il tempo di debounce e il ritardo di debounce
buttonState = False
lastButtonState = False
lastDebounceTime = 0
debounceDelay = 0.05  # 50 millisecondi

# Avvia un iteratore per leggere i pin digitali
it = util.Iterator(board)
it.start()

while True:
    # Leggi lo stato del pulsante
    currentTime = time.time()
    reading = button_pin.read()

    # Verifica se lo stato del pulsante è cambiato
    if reading != lastButtonState:
        lastDebounceTime = currentTime

    # Se il tempo di debounce è trascorso, aggiorna lo stato del pulsante
    if (currentTime - lastDebounceTime) > debounceDelay:
        buttonState = reading

    # Stampa lo stato del pulsante
    print(f'Button state: {buttonState}')

    # Aggiorna lo stato precedente del pulsante
    lastButtonState = reading

    # Aggiungi un ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.01)
```

## POTENZIOMETRO CON LED / POTENTIOMETER WITH LEDs

```python
import time
from pyfirmata import Arduino, util

# Definisci il numero del pin analogico e il numero di LED
analogPin = 0  # A0
ledCount = 5  # Numero di LED
ledPins = [2, 3, 4, 5, 6]  # Array dei pin a cui sono collegati i LED

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Configura i pin dei LED come output
for pin in ledPins:
    board.digital[pin].mode = 1  # OUTPUT

# Configura il pin analogico
analog_pin = board.get_pin(f'a:{analogPin}:i')  # Pin analogico A0 come input

# Avvia un iteratore per leggere i pin analogici
it = util.Iterator(board)
it.start()

# Aggiungi un breve ritardo per consentire l'inizializzazione del board
time.sleep(1)

while True:
    # Leggi il valore del sensore
    sensorReading = analog_pin.read()

    # Verifica che il valore letto non sia None
    if sensorReading is not None:
        # Scala il valore letto (0-1) alla gamma di 0-1023
        sensorReading = sensorReading * 1023

        # Mappa il valore del sensore alla gamma di 0 a ledCount
        ledLevel = int(sensorReading / 1023 * ledCount)

        # Accendi o spegni i LED in base al livello mappato
        for thisLed in range(ledCount):
            if thisLed < ledLevel:
                board.digital[ledPins[thisLed]].write(1)  # HIGH
            else:
                board.digital[ledPins[thisLed]].write(0)  # LOW

    # Aggiungi un ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.1)
```

## COME MISURARE LA POTENZA DI UN LED / HOW TO MEASURE AN LED POWER

```python
import time
from pyfirmata import Arduino, util

# Definisci i numeri dei pin analogici
pin0 = 0  # A0
pin1 = 1  # A1

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Configura i pin analogici
pin_analog0 = board.get_pin(f'a:{pin0}:i')  # Pin analogico A0 come input
pin_analog1 = board.get_pin(f'a:{pin1}:i')  # Pin analogico A1 come input

# Avvia un iteratore per leggere i pin analogici
it = util.Iterator(board)
it.start()

# Aggiungi un breve ritardo per consentire l'inizializzazione del board
time.sleep(1)

while True:
    # Leggi i valori analogici
    value0 = pin_analog0.read()
    value1 = pin_analog1.read()

    # Verifica che i valori letti non siano None
    if value0 is not None and value1 is not None:
        # Scala i valori letti (0-1) alla tensione (0-5V)
        voltaggio0 = value0 * 5.0
        voltaggio1 = value1 * 5.0

        # Calcola la tensione sulla resistenza, la corrente e la potenza
        vr = voltaggio1 - voltaggio0
        corrente = vr / 220.0
        potenza = corrente * voltaggio1

        # Stampa i risultati
        print(f"Corrente: {corrente:.3f} A, Voltaggio: {voltaggio1:.3f} V, Potenza: {potenza:.3f} W")

    # Aggiungi un ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.1)
```

## POTENZIOMETRO + MONITOR SERIALE / POTENTIOMETER + SERIAL MONITOR

```python
import time
from pyfirmata import Arduino, util

# Definisci il pin analogico per il potenziometro
pot = 0  # A0

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Configura il pin analogico
pot_pin = board.get_pin(f'a:{pot}:i')  # Pin analogico A0 come input

# Avvia un iteratore per leggere i pin analogici
it = util.Iterator(board)
it.start()

# Aggiungi un breve ritardo per consentire l'inizializzazione del board
time.sleep(1)

while True:
    # Leggi il valore del potenziometro
    pot_value = pot_pin.read()

    # Verifica che il valore letto non sia None
    if pot_value is not None:
        # Scala il valore letto (0-1) alla gamma di 0-1023
        pot_value = int(pot_value * 1023)

        # Ottieni il timestamp corrente in microsecondi
        timestamp = int(time.time() * 1_000_000)

        # Stampa il timestamp e il valore del potenziometro
        print(f"{timestamp} , {pot_value}")

    # Aggiungi un ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.1)
```

## FOTORESISTENZA ACCENDE LED / PHOTORESISTOR LIGHTS UP LEDs

```python
import time
from pyfirmata import Arduino, util

# Definisci i pin analogici e il numero di LED
cap1 = 0  # A0
cap2 = 1  # A1
ledCount = 5  # Numero di LED
ledPins = [2, 3, 4, 5, 6]  # Array dei pin a cui sono collegati i LED

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Configura i pin dei LED come output
for pin in ledPins:
    board.digital[pin].mode = 1  # OUTPUT

# Configura i pin analogici
cap1_pin = board.get_pin(f'a:{cap1}:i')  # Pin analogico A0 come input
cap2_pin = board.get_pin(f'a:{cap2}:i')  # Pin analogico A1 come input

# Avvia un iteratore per leggere i pin analogici
it = util.Iterator(board)
it.start()

# Aggiungi un breve ritardo per consentire l'inizializzazione del board
time.sleep(1)

while True:
    # Leggi i valori dai pin analogici
    value1 = cap1_pin.read()
    value2 = cap2_pin.read()

    # Verifica che i valori letti non siano None
    if value1 is not None and value2 is not None:
        # Scala i valori letti (0-1) alla gamma di 0-1023
        value1 = value1 * 1023
        value2 = value2 * 1023

        # Calcola la differenza tra i valori letti
        diff = value1 - value2

        # Mappa la differenza alla gamma di 0 a ledCount
        ledLevel = int(diff / 600 * ledCount)
        ledLevel = max(0, min(ledLevel, ledCount))  # Assicura che ledLevel sia tra 0 e ledCount

        # Accendi o spegni i LED in base al livello mappato
        for thisLed in range(ledCount):
            if thisLed < ledLevel:
                board.digital[ledPins[thisLed]].write(1)  # HIGH
            else:
                board.digital[ledPins[thisLed]].write(0)  # LOW

    # Aggiungi un ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.1)
```

## FOTORESISTENZA + MONITOR SERIALE / PHOTORESISTOR + SERIAL MONITOR

```python
import time
from pyfirmata import Arduino, util

# Definisci i pin analogici
cap1 = 0  # A0
cap2 = 1  # A1

# Inizializza la connessione con l'Arduino (sostituisci 'COM7' con la tua porta seriale corretta)
board = Arduino('COM7')

# Configura i pin analogici
cap1_pin = board.get_pin(f'a:{cap1}:i')  # Pin analogico A0 come input
cap2_pin = board.get_pin(f'a:{cap2}:i')  # Pin analogico A1 come input

# Avvia un iteratore per leggere i pin analogici
it = util.Iterator(board)
it.start()

# Aggiungi un breve ritardo per consentire l'inizializzazione del board
time.sleep(1)

while True:
    # Leggi i valori dai pin analogici
    value1 = cap1_pin.read()
    value2 = cap2_pin.read()

    # Verifica che i valori letti non siano None
    if value1 is not None and value2 is not None:
        # Scala i valori letti (0-1) alla gamma di 0-1023
        value1 = value1 * 1023
        value2 = value2 * 1023

        # Calcola i voltaggi
        voltaggio1 = value1 * 5.0 / 1024
        voltaggio2 = value2 * 5.0 / 1024

        # Calcola la differenza di voltaggio
        diff = voltaggio1 - voltaggio2

        # Stampa i risultati
        print(f"{voltaggio1:.2f} V , {voltaggio2:.2f} V , {diff:.2f} V")

    # Aggiungi un ritardo per evitare di sovraccaricare la CPU
    time.sleep(0.5)
```
